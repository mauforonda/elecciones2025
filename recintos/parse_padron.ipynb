```python
import pandas as pd
```

```python
# Descargado de https://www.oep.org.bo/wp-content/uploads/2025/07/ESTADISTICAS_NACIONAL_EG_2025_OFICIAL_.xlsx
PADRON = "recursos/ESTADISTICAS_NACIONAL_EG_2025_OFICIAL_.xlsx"


def parse():
    sheets = {
        "Hab_Inhab_X_Mesa": {
            "columnas": {
                "dep": "departamento_codigo",
                "Departamento": "departamento",
                "prov": "provincia_codigo",
                "Provincia": "provincia",
                "sec": "municipio_codigo",
                "Municipio": "municipio",
                "idloc": "idloc",
                "Asiento Electoral": "asiento",
                "dist": "distrito_codigo",
                "Distrito": "distrito",
                "zon": "zona_codigo",
                "Zona": "zona",
                "Cincun": "cincunscripcion_codigo",
                "Tipo Circunscripcion": "circunscripcion_tipo",
                "reci": "recinto_codigo",
                "Recinto": "recinto",
                "Numero de Mesa": "mesa_codigo",
                "mesa": "mesa",
                "Habilitados ": "habilitados",
                "Inhabilitados": "inhabilitados",
            },
            "skip_rows": 4,
        },
        "Hab_X_mesa_X_grupo Etario": {
            "columnas": {
                "nummesa": "mesa_codigo",
                "Edad18_20": "habilitados_edad_18_20",
                "Edad21_25": "habilitados_edad_21_25",
                "Edad26_30": "habilitados_edad_26_30",
                "Edad31_35": "habilitados_edad_31_35",
                "Edad36_40": "habilitados_edad_36_40",
                "Edad41_45": "habilitados_edad_41_45",
                "Edad46_50": "habilitados_edad_46_50",
                "Edad51_55": "habilitados_edad_51_55",
                "Edad56_60": "habilitados_edad_56_60",
                "Edad61_65": "habilitados_edad_61_65",
                "Edad66_ade": "habilitados_edad_66_adelante",
            },
            "skip_rows": 5,
        },
    }

    excel = pd.ExcelFile(PADRON)
    mesas, edades = [
        pd.read_excel(
            excel,
            sheet_name=sheet_name,
            skiprows=sheets[sheet_name]["skip_rows"],
            skipfooter=1,
            usecols=sheets[sheet_name]["columnas"].keys(),
        ).rename(columns=sheets[sheet_name]["columnas"])
        for sheet_name in ["Hab_Inhab_X_Mesa", "Hab_X_mesa_X_grupo Etario"]
    ]
    return pd.merge(
        mesas,
        edades,
        on="mesa_codigo",
        how="left",
    )
```

```python
df = parse()
```
